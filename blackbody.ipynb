```python
import matplotlib.pyplot as plt
import numpy as np

c = 2.9979E8
h = 6.626E-34
k = 1.3806E-23
T = 6E3

def BB(x):
  y = 2*h*x**3/c**2/(np.exp(x*h/k/T)-1.0)
  return y

x = np.linspace(1e12,1e15, 100)
y = BB(x)

plt.plot(x,y,color='black')
plt.xscale('log')
plt.yscale('log')
plt.ylim(1e-10,5e-8)
plt.xlim(1e12,1e15)
plt.title("Blackbody Spectrum")
plt.xlabel("Frequency")
plt.ylabel("B")
plt.show()

```
